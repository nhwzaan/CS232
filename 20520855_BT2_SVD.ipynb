```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
img = cv.imread("/content/lee-sung-kyung-1.jpg?w=1200&h=0&q=100&dpr=1&fit=crop&s=kngh29WsMQXJPb0ATK4Cyg", cv.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```python
U, S, V = np.linalg.svd(img)
reconstimg = [np.matrix(U[:, :k]) * np.diag(S[:k]) * np.matrix(V[:k, :]) for k in range(2, 100, 2)]

for i in range(len(reconstimg)):
  plt.imshow(reconstimg[i], cmap='gray')
  title = "k = %s" % (i * 2 + 2)
  plt.title(title)
  plt.show()
```
