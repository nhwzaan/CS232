# Import thư viện

```python
import numpy as np
import matplotlib.pyplot as plotter
import matplotlib.pyplot as plt
```

# Preparation

```python
# How many time points are needed i,e., Sampling Frequency
samplingFrequency = 100;

# At what intervals time points are sampled
samplingInterval = 1 / samplingFrequency;

# Begin time period of the signals
beginTime = 0;

# End time period of the signals
endTime = 10; 
```

# Thử nghiệm 1: 2 tín hiệu với hai tần số khác nhau và biên độ khác nhau (độ lệch pha đều bằng 0).

```python
# Frequency of the signals
signal1Frequency = 4;
signal2Frequency = 7;

# Time points
time = np.arange(beginTime, endTime, samplingInterval);


# Create two sine waves
amplitude1 = 2*np.sin(2*np.pi*signal1Frequency*time)
amplitude2 = np.sin(2*np.pi*signal2Frequency*time)

# Create subplot
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4, 1, figsize = (10, 10))
 

# Time domain representation for sine wave 1
ax1.set_title('Sine wave with a frequency of 4 Hz')
ax1.plot(time, amplitude1)
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')

 

# Time domain representation for sine wave 2
ax2.set_title('Sine wave with a frequency of 7 Hz')
ax2.plot(time, amplitude2)
ax2.set_xlabel('Time')
ax2.set_ylabel('Amplitude')


# Add the sine waves
amplitude = amplitude1 + amplitude2

 
# Time domain representation of the resultant sine wave
ax3.set_title('Sine wave with multiple frequencies')
ax3.plot(time, amplitude)
ax3.set_xlabel('Time')
ax3.set_ylabel('Amplitude')


# Frequency domain representation
fourierTransform = np.fft.fft(amplitude)/len(amplitude)           # Normalize amplitude
fourierTransform = fourierTransform[range(int(len(amplitude)/2))] # Exclude sampling frequency

 
tpCount     = len(amplitude)
values      = np.arange(int(tpCount/2))
timePeriod  = tpCount/samplingFrequency
frequencies = values/timePeriod

# Frequency domain representation
ax4.set_title('Fourier transform depicting the frequency components')
ax4.plot(frequencies, abs(fourierTransform))
ax4.set_xlabel('Frequency')
ax4.set_ylabel('Amplitude')
ax4.set_xlim(0,10)


plotter.show()
```

Thử kết hợp lại các sóng từ bảng thành phần được phân tích từ FFT

```python
new_amplitude1=np.sin(2*np.pi*4*time)
new_amplitude2=0.5*np.sin(2*np.pi*7*time)
new_amplitude = new_amplitude1 + new_amplitude2
```

```python
fig, (ax1,ax2) = plt.subplots(2, 1, figsize = (10, 8))
ax1.set_title('New sine wave with multiple frequencies')

ax1.plot(time, new_amplitude)

ax1.set_xlabel('Time')

ax1.set_ylabel('Amplitude')
ax2.set_title('Old sine wave with multiple frequencies')

ax2.plot(time, amplitude)

ax2.set_xlabel('Time')

ax2.set_ylabel('Amplitude')
```


# Thử nghiệm 2: 4 tín hiệu với hai tần số khác nhau và biên độ khác nhau (độ lệch pha đều bằng 0).

```python
# Frequency of the signals
signal1Frequency     = 4;
signal2Frequency     = 7;
signal3Frequency     = 5;
signal4Frequency     = 8;

# Time points
time        = np.arange(beginTime, endTime, samplingInterval);


# Create 4 sine waves
amplitude1 = 2*np.sin(2*np.pi*signal1Frequency*time)
amplitude2 = np.sin(2*np.pi*signal2Frequency*time)
amplitude3 = 3*np.sin(2*np.pi*signal3Frequency*time)
amplitude4 = 4*np.sin(2*np.pi*signal4Frequency*time)

# Create subplot
fig, (ax1,ax2,ax3,ax4,ax5,ax6) = plt.subplots(6, 1, figsize = (15, 17))

 
# Time domain representation for sine wave 1
ax1.set_title('Sine wave with a frequency of 4 Hz')
ax1.plot(time, amplitude1)
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')
 

# Time domain representation for sine wave 2
ax2.set_title('Sine wave with a frequency of 7 Hz')
ax2.plot(time, amplitude2)
ax2.set_xlabel('Time')
ax2.set_ylabel('Amplitude')

# Time domain representation for sine wave 3
ax3.set_title('Sine wave with a frequency of 5 Hz')
ax3.plot(time, amplitude3)
ax3.set_xlabel('Time')
ax3.set_ylabel('Amplitude')


# Time domain representation for sine wave 4
ax4.set_title('Sine wave with a frequency of 8 Hz')
ax4.plot(time, amplitude4)
ax4.set_xlabel('Time')
ax4.set_ylabel('Amplitude')

# Add the sine waves
amplitude = amplitude1 + amplitude2 + amplitude3 +amplitude4
 

# Time domain representation of the resultant sine wave
ax5.set_title('Sine wave with multiple frequencies')
ax5.plot(time, amplitude)
ax5.set_xlabel('Time')
ax5.set_ylabel('Amplitude')


# Frequency domain representation
fourierTransform = np.fft.fft(amplitude)/len(amplitude)           # Normalize amplitude
fourierTransform = fourierTransform[range(int(len(amplitude)/2))] # Exclude sampling frequency


tpCount     = len(amplitude)
values      = np.arange(int(tpCount/2))
timePeriod  = tpCount/samplingFrequency
frequencies = values/timePeriod 

# Frequency domain representation
ax6.set_title('Fourier transform depicting the frequency components')
ax6.plot(frequencies, abs(fourierTransform))

ax6.set_xlabel('Frequency')

ax6.set_ylabel('Amplitude')
ax6.set_xlim(0,10)

 

plotter.show()
```

Thử kết hợp lại các sóng từ bảng thành phần được phân tích từ fft:

```python
new_amplitude1=np.sin(2*np.pi*4*time)
new_amplitude2=0.5*np.sin(2*np.pi*7*time)
new_amplitude3=1.5*np.sin(2*np.pi*5*time)
new_amplitude4=2*np.sin(2*np.pi*8*time)
new_amplitude = new_amplitude1 + new_amplitude2 + new_amplitude3 +new_amplitude4
```

```python
fig, (ax1,ax2) = plt.subplots(2, 1, figsize = (10, 8))
ax1.set_title('New sine wave with multiple frequencies')
ax1.plot(time, new_amplitude)

ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')


ax2.set_title('Old sine wave with multiple frequencies')
ax2.plot(time, amplitude)

ax2.set_xlabel('Time')

ax2.set_ylabel('Amplitude')
```
